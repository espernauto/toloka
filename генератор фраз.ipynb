```python
phrase_list = []
with open('phrases.txt', 'r')  as phrases:
    for k in phrases:
        phrase_list.append(k)
phrases.close()
```

```python
print(phrase_list)
```

```python
import csv
import random
l = []
```

```python
print(len(phrase_list))
```

```python
with open('toloka_send.tsv', 'w') as out_file:
    out_file.write('INPUT:text')
    out_file.write('\n')
    for i in range(10):
        while k in l:
            k = random.choice(phrase_list)
        l.append(k)
        out_file.write(k + '\n')
    print('pool number', int(len(l)/10))
```
