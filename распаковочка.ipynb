```python
import numpy as np

import pandas

data = pandas.read_csv('assignments_from_pool_31145688__08-02-2022.tsv', sep='\t')
needed_data = data.iloc[:, 0:2]
print(needed_data)
```

```python
s = []
with open('phrases.txt', 'r') as f:
    for line in f:
        s.append(line)
```

```python
import shutil
import os
```

```python
l = []
i = 1
for phrase in s:
    j = 0
    os.mkdir(str(i)) #создаем папку
    l.append(str(i) + phrase) #сохраняем связь названия папки и фразы
    my_choice = needed_data.loc[data['INPUT:text'] == phrase.split('\n')[0]]
    for ident in my_choice['OUTPUT:audio_record']: #по идентификатору находим все нужные аудиозаписи и переносим в папку
        for path, subdirs, files in os.walk(os.getcwd()): 
            for filename in files:
                if filename == ident + '.wav' :
                    source_path = path +  '/' + filename
                    dtc =  str(i) + '/' + filename
                    shutil.move(source_path, dtc)
                    
    i+=1
                    
```

```python
print(l)
```

```python
with open('key.txt', 'w', ) as f:
    for phrase in l:
        f.write(phrase + '\n')
```

```python
counter = []
for phrase in s:
    j = 0
    my_choice = needed_data.loc[data['INPUT:text'] == phrase.split('\n')[0]]
    for ident in my_choice['OUTPUT:audio_record']: #по идентификатору находим все нужные аудиозаписи и переносим в папку
        for path, subdirs, files in os.walk(os.getcwd()): 
            for filename in files:
                if filename == ident + '.wav' :
                    j+=1
    counter.append(j)
    if j < 5:
        print(phrase, j)
    if j > 15:
        print(phrase, j)
print(counter)
```

```python
import seaborn as sns
sns_plot = sns.distplot(counter)
fig = sns_plot.get_figure()
```
